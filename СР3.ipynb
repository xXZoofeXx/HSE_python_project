# Задание 3

Имеются данные об изменении численности населения Пермского края.
Год - Родилось - Умерло
2010 - 37304 - 40371
2011 - 37015 - 38685
2012 - 38931 - 37453
2013 - 38637 - 36995
2014 - 38837 - 36901
2015 - 38761 - 37370
2016 - 37258 - 36369
2017 - 31915 - 34861




1.   Постройте двойной линейный график
2.   Добавьте подписи для осей x и y
3.   Скорректируйте отображение значений по осям, если требуется
4.   Добавьте подпись графика



```python
import matplotlib.pyplot as plt

years = ['2010','2011','2012','2013', '2014', '2015', '2016', '2017']
born = [37304, 37015, 38931, 38637, 38837, 38761, 37258, 31915]
died = [40371, 38685, 37453, 36995, 36901, 37370, 36369, 34861]

# 1. Построим двойной график, установим цвета и маркеры
plt.plot(years, born, marker='o', markersize=4, color='green', label='Birth')
plt.plot(years, died, marker='s', markersize=4, color='red', label='Death')

# 2. Добавим подписи осей и легенду
plt.xlabel('Year')
plt.ylabel('People')
plt.legend(loc='upper right', title='Rate:')

# 3. Изменим ротацию значений на оси x на вертикальную и добавим сетку
plt.xticks(rotation='vertical')
plt.grid()

# 4. Добавим заголовок графика
plt.title('Birth/Death Rate')

plt.show() # Вывод результата
```
